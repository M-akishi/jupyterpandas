```python
df = pd.read_csv('clientes_banco.csv')
df.head() 
```

```python
# Eliminar filas duplicadas
df = df.drop_duplicates()

# Rellenar valores nulos en columnas categ√≥ricas
df['Genero'] = df['Genero'].fillna('No informado')

df['Genero'] = df['Genero'].replace('Masculino','M')
df['Genero'] = df['Genero'].replace('Femenino','F')
df['Genero'] = df['Genero'].replace('No informado','N')

df = df.fillna(0.0)

df.head()

df.to_csv('clientes_banco_limpio.csv', index=False)
```

```python
url = "http://127.0.0.1:8000/api/clientes/"

headers= {
    "Accept": "application/json",
    "Content-Type": "application/json"
}

df = pd.read_csv('clientes_banco_limpio.csv')
for idx, data in df.iterrows():
    payload = {
        "Cliente_id": data['Cliente_ID'],
        "Edad": data['Edad'],
        "Saldo": data['Saldo'],
        "Activo": int(data['Activo']),
        "Genero": data['Genero'],
        "Nivel_de_Satisfaccion": int(data['Nivel_de_Satisfaccion']),
    }
    response = requests.post(url, json=payload, headers=headers)
    
    # Maneja la respuesta
    if response.status_code == 201:
        print(f"Cliente {data['Cliente_ID']} creado correctamente.")
    else:
        print(f"Error al crear el cliente {data['Cliente_ID']}: {response.status_code}, {response.text}")

```
